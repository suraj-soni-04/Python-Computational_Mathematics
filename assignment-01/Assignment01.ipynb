```python
import numpy as np
import matplotlib.pyplot as plt
def lagrange_interpolation(x, y, x_val):
  n = len(x)
  p = 0
  for i in range(n):
    term = 1
    for j in range(n):
      if i != j:
        term *= (x_val - x[j]) / (x[i] - x[j])
    p += y[i] * term
  return p
temperature = np.array([0, 20, 40, 60, 80, 100])
pressure = np.array([1.000, 0.889, 0.745, 0.582, 0.412, 0.248])
data = input("Enter tempreature at which pressurebto be calculated in the list separated by space: ")  
temperatures_to_interpolate = data.split()  
temperatures_to_interpolate = [float(temp) for temp in temperatures_to_interpolate]  
pressures_interpolated = lagrange_interpolation(temperature, pressure, temperatures_to_interpolate)
 
print("Calculated interpolated pressures")
print(pressures_interpolated)
 
plt.plot(temperature, pressure, 'o', label='Data points')
plt.plot(temperatures_to_interpolate, pressures_interpolated, '-', label='Interpolated')
plt.xlabel('Temperature (°C)')
plt.ylabel('Pressure (atm)')
plt.title('Lagrange Interpolation for Pressure-Temperature Relationship')
plt.legend()
plt.grid(True)
plt.show()
```
